```python

flights= pd.read_csv('flights.csv')
planes= pd.read_csv('planes.csv')
airlines= pd.read_csv('airlines.csv')
airports=pd.read_csv('airports.csv')
```

# 1.what is the most popular destination city from NewYork?


```python
flights.columns
```

```python
table1 = flights.groupby('dest').agg(count = ('dest','count')).sort_values('count',ascending=False).reset_index()
```

```python
pd.merge(table1,airports[['faa','name']],how = 'left',left_on = 'dest',right_on ='faa')
```

# 3.) which airline is the most punctual

```python
flights['total_delay'] = flights['arr_delay']+flights['dep_delay']
```

```python
table2 = flights.groupby('carrier').agg(mean_delay = ('total_delay','mean')).sort_values('mean_delay').reset_index()
# close to 0 is the most punctual
```

```python
pd.merge(table2,airlines,how ='left',)
```

# 4.) what destination is the longest duration

```python
flights.air_time.describe
```

```python
table3 = flights.groupby(['origin','dest']).agg(average_air_time = ('air_time','mean')).reset_index().sort_values('average_air_time',ascending=False)
```

```python
pd.merge(table3,airports[['faa','name']],how='left',left_on='dest',right_on='faa')
```

# 6.) which airline has the highest capacity of seats

```python
carrier_tailnum = flights[['carrier','tailnum']].drop_duplicates()
seats = pd.merge(carrier_tailnum,planes[['tailnum','seats']],how='left').groupby('carrier').agg(total_seats=('seats','sum')).sort_values('total_seats',ascending=False)
```

# 7.) which airplane model is the highest in use and from which manufacturer

```python
airplane_use = flights.groupby('tailnum').agg(count=('tailnum','count')).reset_index()
```

```python
pd.merge(planes[['tailnum','model','manufacturer']],airplane_use).groupby(['model','manufacturer']).agg(total_flights = ('count','sum')).sort_values('total_flights')
```
